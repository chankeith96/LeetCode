[56. Merge Intervals](https://leetcode.com/problems/merge-intervals/description/)

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # Solution 1:
        # - create result array
        # - sort intervals by start index
        # - iterate through all intervals
        # - Check if overlap with previous interval
        #       - if yes, then merge
        #       - if no, then add previous interval to result
        # - return result
        # Time: O(nlogn)
        # Space: O(n) 

        result = []
        intervals.sort(key = lambda i: i[0])
        prev_interval = intervals[0]

        # edge case, only 1 interval in intervals
        if len(intervals) == 1:
            return intervals

        for interval in intervals[1:]:
            # case: no overlap
            if interval[0] > prev_interval[1] or interval[1] < prev_interval[0]: 
                result.append(prev_interval)
                prev_interval = interval
            # case: overlap
            else:
                prev_interval = [min(prev_interval[0], interval[0]), max(prev_interval[1], interval[1])]
        
        if prev_interval == interval:
            result.append(interval)
        else:
            result.append(prev_interval)

        return result

            
```
