##### Learnings
- Solution 1 was hard to implement because there are many edge cases to consider. Try not to solve Linked List problems by storing nodes in array
- Solution 1.5 also has some edge cases to consider.
- Solution 2 is similar to Solution 3, except that it initializes slow and fast to head instead of dummy node. Therefore, I have to handle edge case where first node is removed. 
- Solution 3 is the best solution by far since don't have to handle any edge cases, all bc it uses dummy node. 
- MAIN LESSON: Use dummy node, especially if the head of the linked list may change.
