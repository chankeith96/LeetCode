[125. Valid Palindrome](https://leetcode.com/problems/valid-palindrome/)

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        # # Solution 1: 
        # # - use list comprehension to filter out non-alphanumeric chars, O(n)
        # # - check if forward pass == backward pass, O(n)
        # # Time: O(n)
        # # Space: O(n)
        # alnum_chars = [char.lower() for char in s if char.isalnum()]
        # return alnum_chars == alnum_chars[::-1]
        
        # Solution 2: Two pointers
        # - use a left pointer for the forward pass, right pointer for backward pass
        # - check alphanumeric chars equivalent, skip every non-alphanumeric char
        # - go until pointers cross
        # Time: O(n)
        # Space: O(1)
        left_ptr = 0
        right_ptr = len(s) - 1

        while left_ptr < right_ptr:

            while left_ptr < right_ptr and not s[left_ptr].isalnum():
                left_ptr += 1
            while left_ptr < right_ptr and not s[right_ptr].isalnum():
                right_ptr -= 1
            if s[left_ptr].lower() == s[right_ptr].lower():
                left_ptr += 1
                right_ptr -= 1
            else:
                return False
        return True
```
