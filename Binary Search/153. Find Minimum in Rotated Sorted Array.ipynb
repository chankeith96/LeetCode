[153. Find Minimum in Rotated Sorted Array](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/)

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        # Solution 0: Obvious brute force method is to iterate through array and find min  
        # Time: O(n)
        # Space: O(1)

        # # Solution 1: Binary Search for min using recursion
        # # - take the middle value at index mid
        # # - if mid < mid-1 AND mid > mid+1, then return mid
        # # - if mid-1 < mid+1, then check left
        # # - else check right
        # # Time: O(logn)
        # # Space: O(logn) bc recursive call stack?
        # left = 0
        # right = len(nums) - 1
        # mid = left + (right - left) // 2

        # # Base Case
        # if len(nums) == 1:
        #     return nums[0]
        # if nums[mid] < nums[mid-1] and nums[mid] < nums[mid+1]:
        #     return nums[mid]
        # if nums[mid-1] < nums[right]:)
        #     return self.findMin(nums[left:mid])
        # else:
        #     print(nums[mid+1:right+1])
        #     return self.findMin(nums[mid+1:right+1])

        # Solution 2: Binary Search for min
        # Time: O(logn)
        # Space: O(1)
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = left + (right - left) // 2
            if left == right:
                return nums[left]
            if nums[mid] < nums[mid-1] and nums[mid] < nums[mid+1]:
                return nums[mid]
            if nums[mid] < nums[right]:
                right = mid - 1
            else:
                left = mid + 1
```

##### Edge Cases
- Recognize that there are only 4 cases for the rotation
    - rotated n times (back to original)
    - min value at the middle
    - min value on the left
    - min value on the right
